```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data

```python
customers_df = pd.read_csv("data/olist_customers_dataset.csv")
sellers_df = pd.read_csv("data/olist_sellers_dataset.csv")
orderreviews_df = pd.read_csv("data/olist_order_reviews_dataset.csv")
orderitems_df = pd.read_csv("data/olist_order_items_dataset.csv")
products_df = pd.read_csv("data/olist_products_dataset.csv")
geolocations_df = pd.read_csv("data/olist_geolocation_dataset.csv")
categoryname_df = pd.read_csv("data/product_category_name_translation.csv")
orders_df = pd.read_csv("data/olist_orders_dataset.csv")
payments_df = pd.read_csv("data/olist_order_payments_dataset.csv")
```

```python
df= pd.merge(orders_df, customers_df, on='customer_id', how='inner')
df= pd.merge(df, orderreviews_df, on='order_id', how='inner')
df= pd.merge(df, orderitems_df, on='order_id', how='inner')
df= pd.merge(df, payments_df, on='order_id', how='inner')
df= pd.merge(df, products_df, on='product_id', how='inner')
df= pd.merge(df, sellers_df, on='seller_id', how='inner')
df= pd.merge(df, categoryname_df, on='product_category_name', how='inner')
df
```

```python
# Drop None rows
df.dropna(subset= ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date'], inplace=True)

# Drop row 10342 since 4 missing values are in this row
df.drop(10342, inplace=True)
df.reset_index(inplace= True, drop= True)

# Ignore the missing values in review_comment_title and review_comment_message since we don't need them in this notebook

df.isna().sum()
```

```python
# Convert all the datetime columns in to pandas datetime format
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_approved_at'] = pd.to_datetime(df['order_approved_at'])
df['order_delivered_carrier_date'] =pd.to_datetime(df['order_delivered_carrier_date'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])
df['review_creation_date'] = pd.to_datetime(df['review_creation_date'])
df['review_answer_timestamp'] = pd.to_datetime(df['review_answer_timestamp'])
df['shipping_limit_date'] = pd.to_datetime(df['shipping_limit_date'])
df
```

## Labeling

```python
# Using all the dattime columns to calcualte the times btween
df['approved_hours'] = (df['order_approved_at'] - df['order_purchase_timestamp']).dt.seconds / 3600
df['estimated_days'] = (df['order_estimated_delivery_date'] - df['order_purchase_timestamp']).dt.days
df['arrival_days'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
df['shipping_days'] = (df['order_delivered_customer_date'] - df['order_delivered_carrier_date']).dt.days
df['carrier_status'] = (df['shipping_limit_date'] - df['order_delivered_carrier_date']).dt.days.apply(lambda x : 'Early' if x >=0 else 'Late')
df['arrival_status'] = (df['order_estimated_delivery_date'] - df['order_delivered_customer_date']).dt.days.apply(lambda x : 'Early' if x >=0 else 'Late')
df['review_answering'] = (df['review_answer_timestamp'] - df['review_creation_date']).dt.seconds / 3600
df
```

```python
# Categorize the time difference using qcut
df['approved_speed'] = pd.qcut(df['approved_hours'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Very Fast', 'Fast', 'Slow', 'Very Slow'])
df['delivery_speed'] = pd.qcut(df['estimated_days'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Very Fast', 'Fast', 'Slow', 'Very Slow'])
df['arrival_speed'] = pd.qcut(df['arrival_days'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Very Fast', 'Fast', 'Slow', 'Very Slow'])
df['shipping_speed'] = pd.qcut(df['shipping_days'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Very Fast', 'Fast', 'Slow', 'Very Slow'])
df['review_answering'] = pd.qcut(df['review_answering'], q=[0, 0.25, 0.5, 0.75, 1], labels=['Very Fast', 'Fast', 'Slow', 'Very Slow'])
df
```

### calculate the distance between seller and customer based on longitude and latitude

```python
geo_distance = df[['customer_city', 'freight_value','seller_city', 'category_label']].copy()

# Create a dictionary mapping cities to geolocation_lat and geolocation_lng values
geo_dict = dict(zip(geolocations_df['geolocation_city'], geolocations_df[['geolocation_lat', 'geolocation_lng']].values))

# Map locations to dataframe
geo_distance['customer_lat'], geo_distance['customer_lng'] = zip(*[(geo_dict[i][0], geo_dict[i][1]) if i in geo_dict else (np.nan, np.nan) for i in geo_distance['customer_city']])
geo_distance['seller_lat'], geo_distance['seller_lng'] = zip(*[(geo_dict[i][0], geo_dict[i][1]) if i in geo_dict else (np.nan, np.nan) for i in geo_distance['seller_city']])

geo_distance
```

```python
geo_distance.isna().sum()
```

```python
geo_distance.dropna(inplace=True)
geo_distance
```

```python
geo_distance.isna().sum()
```

```python
import geopy.distance

# calculate the distance between seller and customer based on longitude and latitude
geo_distance['distance'] = geo_distance.apply(lambda row: geopy.distance.distance((row['seller_lat'],row['seller_lng']), (row['customer_lat'], row['customer_lng'])).km, axis=1)

# plot the distribution of distances using a histogram
plt.figure(figsize=(10,8))
sns.histplot(x="distance", data=geo_distance, bins=50, kde=True)
plt.title("Distribution of Distance between Sellers and Customers")
plt.xlabel("Distance (km)")
plt.ylabel("Count")
plt.show()
```
